```python
from keras.models import Sequential
from keras.layers import Dense
import keras
import pandas as pd
import numpy as np

mydata = pd.read_csv("SpiralWithCluster.csv")


x = mydata[["x","y"]].values
y = mydata["SpectralCluster"].values

print(mydata.head())
print(mydata.shape)
```

```python
##SIGMOID
model_sigmoid = Sequential()
model_sigmoid.add(Dense(6, input_dim=2, activation='sigmoid'))
model_sigmoid.add(Dense(6, activation='sigmoid'))
model_sigmoid.add(Dense(6, activation='sigmoid'))
model_sigmoid.add(Dense(1, activation='sigmoid'))

model_sigmoid.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model_sigmoid.fit(x, y, epochs=2000, batch_size=1)

_, accuracy_sigmoid = model_sigmoid.evaluate(x, y)
print('Accuracy: %.2f' % (accuracy_sigmoid*100))
predictions_sigmoid = model_sigmoid.predict_classes(x)
```

```python
##RELU
model_relu = Sequential()
model_relu.add(Dense(7, input_dim=2, activation='relu'))
model_relu.add(Dense(7, activation='relu'))
model_relu.add(Dense(7, activation='relu'))
model_relu.add(Dense(7, activation='relu'))
model_relu.add(Dense(1, activation='sigmoid'))
model_relu.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model_relu.fit(x, y, epochs=500, batch_size=1)

_, accuracy_relu = model_relu.evaluate(x, y)
print('Accuracy: %.2f' % (accuracy_relu*100))
predictions_relu = model_relu.predict_classes(x)
```

```python
##TANH

model_tanh = Sequential()
model_tanh.add(Dense(9, input_dim=2, activation='tanh'))
model_tanh.add(Dense(9, activation='tanh'))
model_tanh.add(Dense(9, activation='tanh'))
model_tanh.add(Dense(1, activation='sigmoid'))
model_tanh.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model_tanh.fit(x, y, epochs=500, batch_size=1)

_, accuracy_tanh = model_tanh.evaluate(x, y)
print('Accuracy: %.2f' % (accuracy_tanh*100))
predictions_tanh = model_tanh.predict_classes(x)
```
